```python
images = df_test.iloc[:, 1:].values  
labels = df_test['label'].values
```

```python
num_images= 35
num_columns = 5  
num_rows = num_images // num_columns  

plt.figure(figsize=(15, 12))

for i in range(num_images):
    plt.subplot(num_rows, num_columns, i+1)
    img = images[i].reshape(28, 28)  # Size image 28x28
    plt.imshow(img, cmap='gray')
    plt.title(f"Label: {labels[i]}", fontsize=10)
    plt.axis('off')
```

```python
#image label 0
label_0_data = df_test[df_test['label'] == 0]

num_images_to_display = 15

num_columns_0 = 5  
num_rows_0 = num_images_to_display // num_columns_0

label_0_images = label_0_data.iloc[:num_images_to_display, 1:].values

plt.figure(figsize=(15, 10))

for i in range(num_images_to_display):
    plt.subplot(num_rows_0, num_columns_0, i+1)
    img = label_0_images[i].reshape(28, 28)  # image size 28x28
    plt.imshow(img, cmap='gray')
    plt.title(f"Label: 0")
    plt.axis('off')
```
